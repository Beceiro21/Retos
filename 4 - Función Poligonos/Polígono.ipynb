# Area de polígonos.
 * Crea una única función (importante que sólo sea una) que sea capaz de calcular y retornar el área de un polígono.
   - La función recibirá por parámetro sólo UN polígono a la vez.
   - Los polígonos soportados serán Triángulo, Cuadrado y Rectángulo.
   - Imprime el cálculo del área de un polígono de cada tipo.


```python
# Importamos las libreías necesarias
from math import pi

# Creamos la función
def areaPoligonos():

# Realizamos la pregunta inicial
    case = input('De los siguientes polígonos, cuadrado, círculo, triángulo y rectángulo, de cúal quieres saber su área?').lower()

# Creamos una función anidada para interactuar con el usuario con el argumento case, que es la respuesta del usuario
    def swich_case(case):

# Con if, elif y else controlamos el flujo la información      
        if case == 'circulo':
            r = int(input('De qué radio es el circulo?'))
            areaCirculo = pi * r**2
            # Redondeamos decimales
            circulo =  f"{areaCirculo:.3f}"
            print('El área del', case, 'es: ', circulo)
            
        elif case == 'triangulo':
            base = int(input('De cuanto es la base?'))
            altura = int(input('Y la altura?'))
            areaTriangulo = (base*altura)/2
            print('El área del', case, 'es: ', areaTriangulo)

        elif case == 'cuadrado':
            l1 = int(input('De cuanto es el primer lado?'))
            l2 = int(input('Y el segundo?'))
            areaCuadrado = l1*l2
            print('El área del', case, 'es: ',areaCuadrado)

        elif case == 'rectangulo':
            largo = int(input('De cuanto es el largo?'))
            ancho = int(input('Y el ancho?'))
            areaRectangulo = largo*ancho
            print('El área del', case, 'es: ', areaRectangulo)

        else:
            print('Por favor, introduce uno de los elementos indicados')
# Imprimimos la función anidada    
    swich_case(case)

areaPoligonos()
```
