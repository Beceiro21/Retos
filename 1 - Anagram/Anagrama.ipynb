 # Anagrama
 
 
 
 * Escribe una función que reciba dos palabras (String) y retorne verdadero o falso (Bool) según sean o no anagramas.
   - Un Anagrama consiste en formar una palabra reordenando TODAS las letras de otra palabra inicial.
   - NO hace falta comprobar que ambas palabras existan.
   - Dos palabras exactamente iguales no son anagrama.

```python
def anagrama():

    # Pedimos las palabras al usuario y las ponemos en minúsculas para evitar errores.
    primeraPalabra = input('Introduce una palabra').lower()
    segundaPalabra = input('introduce otra palabra').lower()

    # Las convertimos en listas
    primeraPalabraLista = list(primeraPalabra)
    segundaPalabraLista = list(segundaPalabra)

    # Ordenamos las listas
    primeraPalabraLista.sort()
    segundaPalabraLista.sort()

    # Con el método ("".join()) lo juntamos todo convirtiendolo en string 
    primer = "".join(primeraPalabraLista)
    segun = "".join(segundaPalabraLista)

    # Pedimos que la función nos devuelva sólamente si es verdadero o no
    return  primer == segun



    

```

```python
# Llamamos a la función
anagrama()
```
