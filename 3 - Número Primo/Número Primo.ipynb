# FizzBuzz


 * Escribe un programa que muestre por consola (con un print) los
 * números de 1 a 100 (ambos incluidos y con un salto de línea entrecada impresión), sustituyendo los siguientes:
   - Múltiplos de 3 por la palabra "fizz".
   - Múltiplos de 5 por la palabra "buzz".
   - Múltiplos de 3 y de 5 a la vez por la palabra "fizzbuzz".


```python
# Iniciamos el bucle del 1 hasta el 100
for i in range(1,101):
# Primero realizamos la pregunta si se cumplen las dos condiciones
    if i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz')
# Y después las siguientes
    elif i % 5 == 0:
        print('buzz')
    elif i % 3 == 0:
        print('fizz')
    else:
        print(i)
```
