# Número Primo
 * Escribe un programa que se encargue de comprobar si un número es o no primo.
 * Hecho esto, imprime los números primos entre 1 y 100.
 * Un número primo solamente se puede dividir entre si mismo o 1.

```python
# Preguntamos al usuario por el número que quiera comprobar
pideNumero = int(input('Inserta un número para saber si es primo o no'))
es_primo = True

# Comprobamos mediante un bucle FOR que nos divida desde 2 hasta el número requerido
for i in range(2,pideNumero):
    # Le realizamos la pregunta que si el resultado es 0
    if  pideNumero % i == 0:
    # eso quiere decir que el número es divisible por más números
        es_primo = False
    # En ese momento rompemos el bucle
        break

# Realizamos la pregunta final sobre si es primo o no
if es_primo:
    print('El número introducido es primo')
else:
    print('El número introducido NO es primo')
```
